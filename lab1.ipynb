```python
img = cv2.imread("lena.png")
h,w,c = img.shape #set h,w,c equals to the image's rows, columns, and channels separately
img2 =  np.zeros([h,w,c], dtype=np.uint8) #create a blank image which has the same properties of the original image
img3 =  np.zeros([h,w,c], dtype=np.uint8)
fig = plt.figure(figsize=(10, 7)) #create figure framework
rows = 2
columns = 2

for y in range(h): #a loop which iterate all the pixels in the image
    for x in range(w//2): #we only need to iterate half of all the pixels
        img2[w-x-1,y] = img[x,y] #copy one pixel of the original pixel and paste it on the other side of the blank pixel(i.e.copy top and paste it at the bottom)
        img2[x,y] = img[w - x - 1,y] #copy the other side of the original pixel and place it at the other side of the blank image(i.e. copy bottom and paste it at the top
        
for y in range(h//2): 
    for x in range(w):
        img3[x,h-y-1] = img[x,y] #copy one pixel of the original pixel and paste it on the other side of the blank pixel(i.e. copy right and paste it at the left)
        img3[x,y] = img[x,h-y-1] #copy one pixel of the original pixel and paste it on the other side of the blank pixel(i.e. copy left and paste it at the right)
        
fig.add_subplot(rows, columns, 1)    #set the first image to be put in the first row and column     
plt.imshow(cv2.cvtColor(img3,cv2.COLOR_BGR2RGB))
plt.title("Lena coclor horizontal flip")

fig.add_subplot(rows, columns, 2)     #set the first image to be put in the first row and the second column      
plt.imshow(cv2.cvtColor(img2,cv2.COLOR_BGR2RGB))
plt.title("Lena Color vertical flip")

fig.add_subplot(rows, columns, 3)     #set the first image to be put in the second row and the first column      
gray3 = cv2.cvtColor(img3,cv2.COLOR_BGR2GRAY)
plt.imshow(gray3, cmap = 'gray')
plt.title("Lena horizontal flip")

fig.add_subplot(rows, columns, 4)     #set the first image to be put in the second row and column
gray2 = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
plt.imshow(gray2, cmap = 'gray')
plt.title("Lena vertical flip")
```

```python
img = cv2.imread("lena.png")
h,w,c = img.shape
img4 =  np.zeros([h,w,c], dtype=np.uint8)

fig = plt.figure(figsize=(10, 7))
rows = 2
columns = 2

for y in range(h):
    for x in range(w):
        pixel = img[x,y]        
        #negate color
        pixel[0] = 255 - pixel[0] 
        pixel[1] = 255 - pixel[1]
        pixel[2] = 255 - pixel[2]
        #modify image with invert colors
        img4[x,y] = pixel

fig.add_subplot(rows, columns, 1)   
plt.imshow(cv2.cvtColor(img4,cv2.COLOR_BGR2RGB))
plt.title("Lena color negated")

fig.add_subplot(rows, columns, 2)
gray4 = cv2.cvtColor(img4,cv2.COLOR_BGR2GRAY)
plt.imshow(gray4, cmap = 'gray')
plt.title("Lena gray negated")
```

```python
img = cv2.imread("lena.png")
h,w,c = img.shape
img5 =  np.zeros([h,w,c], dtype=np.uint8)
img6 =  np.zeros([h,w,c], dtype=np.uint8)
img7 =  np.zeros([h,w,c], dtype=np.uint8)

fig = plt.figure(figsize=(10, 7))
rows = 2
columns = 3

for y in range(h):
    for x in range(w):
        RGB = img[x,y] 
        img5[x,y] = (RGB[0],0,0) #separate image blue
        img6[x,y] = (0,RGB[1],0) #separate image green
        img7[x,y] = (0,0,RGB[2]) #separate image red
        
fig.add_subplot(rows, columns, 1)
plt.imshow(cv2.cvtColor(img5,cv2.COLOR_BGR2RGB))
plt.title("Lena blue image")

fig.add_subplot(rows, columns, 2)
plt.imshow(cv2.cvtColor(img6,cv2.COLOR_BGR2RGB))
plt.title("Lena green image")

fig.add_subplot(rows, columns, 3)
plt.imshow(cv2.cvtColor(img7,cv2.COLOR_BGR2RGB))
plt.title("Lena red image")
```
